```python
import pandas as pd
df=pd.read_csv("SOCR-HeightWeight.csv")
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
h=df['Height(Inches)']
w=df['Weight(Pounds)']
print("MEan H of the class: ",h.mean())
print("Mean W of the class: ",w.mean())
print("Median H of the class: ",h.median())
print("Median W of the class: ",w.median())
print("Mode H  of the class: ",h.mode())
print("Mode W of the class: ",w.mode())
plt.scatter(h,w,color='yellow')
```
